```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("2024_turkey_budget_sum.csv")
```

```python
df['TOPLAM'] = df['TOPLAM'] / 1_000_000
```

```python
df['TOPLAM'] = df['TOPLAM'].astype(int)
```

# Matplotlib Visualization

```python
# 1. Bar Plot (sorted high to low)
df_sorted = df.sort_values('TOPLAM', ascending=False)
plt.figure(figsize=(12, 6))
plt.bar(df_sorted['KURUMLAR'], df_sorted['TOPLAM'])
plt.title('Total Budget by Institution (Sorted High to Low)')
plt.xticks(rotation=90)
plt.ylabel('Budget (in millions)')
plt.tight_layout()
plt.show()
```

```python
# 2. Horizontal Bar Plot
df_sorted = df.sort_values('TOPLAM', ascending=True)
plt.figure(figsize=(10, 8))
plt.barh(df_sorted['KURUMLAR'], df_sorted['TOPLAM'])
plt.title('Total Budget by Institution (Horizontal)')
plt.show()
```

```python
# 3. Pie Chart
plt.figure(figsize=(10, 10))
plt.pie(df['TOPLAM'], labels=df['KURUMLAR'], autopct='%1.1f%%',  )
plt.title('Budget Distribution')
plt.show()
```

```python
# 4. Donut Chart
plt.figure(figsize=(10, 10))
plt.pie(df['TOPLAM'], labels=df['KURUMLAR'], autopct='%1.1f%%', pctdistance=0.85)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Budget Distribution (Donut)')
plt.show()
```

```python
# 5. Lollipop Chart
df_sorted = df.sort_values('TOPLAM', ascending=False)
plt.figure(figsize=(12, 6))
plt.stem(df_sorted['KURUMLAR'], df_sorted['TOPLAM'])
plt.title('Budget by Institution (Lollipop)')
plt.xticks(rotation=90)
plt.show()
```

```python
# 6. Polar Bar Chart
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)
theta = np.linspace(0, 2*np.pi, len(df))
ax.bar(theta, df['TOPLAM'], width=0.3)
ax.set_xticks(theta)
ax.set_xticklabels(df['KURUMLAR'], rotation=45)
plt.title('Budget Distribution (Polar)')
plt.show()
```

```python
# 7. Bubble Chart
sizes = df['TOPLAM'] / 1000
plt.figure(figsize=(12, 8))
plt.scatter(range(len(df)), df['TOPLAM'], s=sizes)
plt.title('Budget by Institution (Bubble)')
plt.xticks(range(len(df)), df['KURUMLAR'], rotation=90)
plt.show()
```

```python
# 8. Area Chart
df_sorted = df.sort_values('TOPLAM', ascending=False)
plt.figure(figsize=(12, 6))
plt.fill_between(range(len(df)), df_sorted['TOPLAM'])
plt.title('Cumulative Budget')
plt.xticks(range(len(df)), df_sorted['KURUMLAR'], rotation=90)
plt.show()
```

```python
# 9. Box Plot
plt.figure(figsize=(8, 6))
plt.boxplot(df['TOPLAM'])
plt.title('Budget Distribution (Box Plot)')
plt.show()
```

```python
# 10. Violin Plot
plt.figure(figsize=(8, 6))
plt.violinplot(df['TOPLAM'])
plt.title('Budget Distribution (Violin Plot)')
plt.show()
```

```python
# 11. Scatter Plot
df_sorted = df.sort_values('TOPLAM', ascending=False)
plt.figure(figsize=(10, 6))
plt.scatter(range(len(df)), df_sorted['TOPLAM'])
plt.title('Budget by Institution (Scatter)')
plt.xticks(range(len(df)), df_sorted['KURUMLAR'], rotation=90)
plt.show()
```

```python
# 12. Step Plot
plt.figure(figsize=(12, 6))
plt.step(range(len(df)), df['TOPLAM'])
plt.title('Budget by Institution (Step)')
plt.xticks(range(len(df)), df['KURUMLAR'], rotation=90)
plt.show()
```

```python
# 13. Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['TOPLAM'], bins=10)
plt.title('Budget Distribution (Histogram)')
plt.show()
```

```python
# 14. Cumulative Plot
plt.figure(figsize=(12, 6))
plt.plot(range(len(df)), df['TOPLAM'].cumsum())
plt.title('Cumulative Budget')
plt.xticks(range(len(df)), df['KURUMLAR'], rotation=90)
plt.show()
```

```python
# 15. Simple Line Plot
plt.figure(figsize=(12, 6))
plt.plot(range(len(df)), df['TOPLAM'], marker='o')
plt.title('Budget by Institution (Line Plot)')
plt.xticks(range(len(df)), df['KURUMLAR'], rotation=90)
plt.show()
```
