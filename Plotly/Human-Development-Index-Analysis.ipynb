# Human Development Index Analysis with Plotly

Dataset: https://www.kaggle.com/datasets/iamsouravbanerjee/human-development-index-dataset

Example Solution: https://thecleverprogrammer.com/2022/12/26/social-progress-index-analysis-using-python/

Hugging Face: https://huggingface.co/spaces/alperugurcan/hdi-visualization

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load the dataset
df = pd.read_csv('Human Development Index - Full.csv')

pd.set_option('display.max_columns', None)

# Display the first few rows of the dataset
df.head()
```

```python
# 'Human Development Index' ile başlayan sütunları seç
hdi_columns = [col for col in df.columns if col.startswith('Human Development Index')]

# Temel bilgi sütunlarını ekle
base_columns = ['ISO3', 'Country', 'Human Development Groups', 'HDI Rank (2021)']

# Tüm seçilmiş sütunları birleştir
columns_to_use = base_columns + hdi_columns
```

```python
df_hdi = df[columns_to_use]
```

```python
df_hdi.head()
```

```python
# Silinecek sütunların listesi
columns_to_delete = [
    'Human Development Index (1990)', 'Human Development Index (1991)',
    'Human Development Index (1992)', 'Human Development Index (1993)',
    'Human Development Index (1994)', 'Human Development Index (1995)',
    'Human Development Index (1996)', 'Human Development Index (1997)',
    'Human Development Index (1998)', 'Human Development Index (1999)',
    'Human Development Index (2000)', 'Human Development Index (2001)',
    'Human Development Index (2002)', 'Human Development Index (2003)',
    'Human Development Index (2004)', 'Human Development Index (2005)',
    'Human Development Index (2006)', 'Human Development Index (2007)',
    'Human Development Index (2008)', 'Human Development Index (2009)'
]

# Sütunları sil
df_hdi = df_hdi.drop(columns=columns_to_delete)

# Kalan sütunları göster
print("Remaining columns:")
print(df_hdi.columns.tolist())

# Veri setinin yeni boyutlarını kontrol et
print(f"\nNew dataset shape: {df_hdi.shape}")

# Kalan HDI sütunlarının eksik değerlerini kontrol et
hdi_columns = [col for col in df_hdi.columns if col.startswith('Human Development Index')]
print("\nMissing values in remaining HDI columns:")
print(df_hdi[hdi_columns].isnull().sum())
```

```python
df_hdi.dropna(inplace=True)
```

```python
del df_hdi['ISO3']
```

```python
del df_hdi['HDI Rank (2021)']
```

```python
df_hdi.head()
```

```python
# Sort the dataframe by 'Human Development Index (2021)' in descending order
df_hdi_sorted = df_hdi.sort_values('Human Development Index (2021)', ascending=False)

# Create a new column 'HDI Rank (2021)' with the rank
df_hdi_sorted['HDI Rank (2021)'] = range(1, len(df_hdi_sorted) + 1)

# Reorder the columns to put 'HDI Rank (2021)' right after 'Country'
columns_order = ['Country', 'HDI Rank (2021)'] + [col for col in df_hdi_sorted.columns if col not in ['Country', 'HDI Rank (2021)']]
df_hdi_sorted = df_hdi_sorted[columns_order]

# Display the first few rows of the updated dataframe
df_hdi_sorted.head()
```

```python
df_years = df_hdi_sorted.drop(columns=['HDI Rank (2021)', 'Human Development Groups'])
df_years.head()
```

```python
df_rank = df_hdi_sorted[['Country', 'HDI Rank (2021)']]
df_rank.head()
```

```python
df_hdi_groups = df_hdi_sorted[['Country', 'Human Development Groups']]
df_hdi_groups.head()
```

# Plotly Visualization


```python
# 1. Bar chart showing the top 10 countries by HDI in 2021
fig1 = px.bar(df_years.head(10), x='Country', y='Human Development Index (2021)', title='Top 10 Countries by HDI (2021)', color='Human Development Index (2021)', color_continuous_scale='Viridis')
fig1.show()
```

```python
# 2. Pie chart showing the distribution of countries by HDI Groups
fig2 = px.pie(df_hdi_groups, names='Human Development Groups', 
              title='Distribution of Countries by HDI Groups')
fig2.show()
```

```python
# 3. Line chart showing HDI trends for the top 5 countries
top_5 = df_rank.head()['Country'].tolist()
fig3 = px.line(df_years[df_years['Country'].isin(top_5)], x=df_years.columns[1:], y='Country', 
               title='HDI Trends for Top 5 Countries')
fig3.show()
```

```python
# 4. Bar chart showing the bottom 10 countries by HDI in 2021
fig4 = px.bar(df_years.tail(10), x='Country', y='Human Development Index (2021)', 
              title='Bottom 10 Countries by HDI (2021)')
fig4.show()
```

```python
# 5. Bar chart showing the top 10 countries with highest HDI improvement from 2010 to 2021
df_years['HDI_change'] = df_years['Human Development Index (2021)'] - df_years['Human Development Index (2010)']
fig5 = px.bar(df_years.nlargest(10, 'HDI_change'), x='Country', y='HDI_change', 
              title='Top 10 Countries with Highest HDI Improvement (2010-2021)')
fig5.show()
```

```python
# 6. Box plot showing the distribution of HDI values in 2021
fig6 = px.box(df_years, y='Human Development Index (2021)', 
              title='Distribution of HDI Values (2021)')
fig6.show()
```

```python
# 7. Choropleth map showing HDI values for 2021
fig7 = px.choropleth(df_years, locations='Country', locationmode='country names', 
                     color='Human Development Index (2021)',
                     title='World Map of Human Development Index (2021)', 
                     color_continuous_scale='Viridis')
fig7.show()
```

```python
# 8. Scatter plot comparing HDI values in 2010 and 2021
fig8 = px.scatter(df_years, x='Human Development Index (2010)', y='Human Development Index (2021)', 
                  hover_name='Country', title='HDI Comparison: 2010 vs 2021')
fig8.add_trace(go.Scatter(x=[0, 1], y=[0, 1], mode='lines', name='No Change Line'))
fig8.show()
```

```python
# 9. Box plot showing HDI distribution by development groups in 2021
fig9 = px.box(df_hdi_sorted, x='Human Development Groups', y='Human Development Index (2021)', 
              title='HDI Distribution by Development Groups (2021)')
fig9.show()
```

```python
# 10. Sunburst chart showing HDI distribution by groups and top countries
df_sunburst = df_hdi_sorted.copy()
df_sunburst['HDI_2021'] = pd.cut(df_sunburst['Human Development Index (2021)'], 
                                 bins=[0, 0.55, 0.7, 0.8, 1], 
                                 labels=['Low', 'Medium', 'High', 'Very High'])

fig10 = px.sunburst(df_sunburst, 
                    path=['HDI_2021', 'Human Development Groups', 'Country'], 
                    values='Human Development Index (2021)',
                    color='HDI_2021',
                    color_discrete_map={'Low': 'red', 'Medium': 'orange', 
                                        'High': 'lightgreen', 'Very High': 'darkgreen'},
                    title='HDI Distribution by Groups and Top Countries (2021)')
fig10.update_traces(textinfo="label+percent entry")
fig10.show()
```
