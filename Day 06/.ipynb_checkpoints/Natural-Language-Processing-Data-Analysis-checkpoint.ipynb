```python
df = pd.read_csv("yelp.csv")
```

```python
df.head()
```

```python
df["text"] = df["text"].str.lower()
```

```python
# punc
```

```python
df['text'] = df['text'].str.replace('\n', '', regex=True) # noktalama işaretleri
df['text'] = df['text'].str.replace('\d+', '', regex=True) #rakam kaldır
df['text'] = df['text'].str.replace('\n', '', regex=True) # ?
df['text'] = df['text'].str.replace('\r', '', regex=True) # ?
```

```python
df["text"][0]
```

```python
#Sentiment Anality - Duygu durum analizi
```

```python
ybw = df[(df.stars==1)|(df.stars==5)]
```

```python
ybw.head() #index bozuldu
```

```python
ybw.reset_index(drop=True, inplace=True)
```

```python
ybw.head()
```

```python
x=ybw[["text"]]
y=ybw[["stars"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
# vectore çevir
vect=CountVectorizer(stop_words="english",ngram_range=1,2)
```

```python
xyeni = vect.fit_transform(x['text'])
```

```python
x_train,x_test,y_train,y_test = train_test_split(xyeni,y,test_size = 0.20, random_state = 42)
```

```python
nb = MultinomialNB()
```

```python
model = nb.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred)
```
