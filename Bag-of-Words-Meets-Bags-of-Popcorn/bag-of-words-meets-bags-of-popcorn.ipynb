# Bag-of-Words-Meets-Bags-of-Popcorn

This Movie Review Sentiment Analysis project uses DistilBERT to automatically classify reviews as positive or negative, deployed as a user-friendly web app on Hugging Face Spaces.

Users can simply paste their review text and instantly get sentiment results, making it a practical tool for quick opinion analysis.

Data: https://www.kaggle.com/competitions/word2vec-nlp-tutorial/data

Hugging Face: https://huggingface.co/spaces/alperugurcan/movie-reviews-sentiment-analysis

```python
import pandas as pd
from transformers import pipeline
```

```python
train = pd.read_csv('/kaggle/input/word2vec-nlp-tutorial/labeledTrainData.tsv.zip', delimiter='\t')
test = pd.read_csv('/kaggle/input/word2vec-nlp-tutorial/testData.tsv.zip', delimiter='\t')
```

```python
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
def get_sentiment(text):
    result = classifier(text[:512])[0]
    return 1 if result['label'] == 'POSITIVE' else 0
```

```python
import swifter
test['sentiment'] = test['review'].swifter.apply(get_sentiment)
```

```python
submission = test[['id', 'sentiment']]
submission.to_csv('submission.csv', index=False)
```
