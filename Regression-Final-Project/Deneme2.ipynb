```python
base_url="https://www.zingat.com/balikesir-kiralik"
houselinks = []

for x in range(1,7):
    req = Request(
        url=(f'https://www.zingat.com/balikesir-kiralik?page={x}'), 
        headers={'User-Agent': 'Mozilla/5.0'}
    )
    
    # Open the URL and read the response
    response = urlopen(req)
    webpage_bytes = response.read()
    webpage_text = webpage_bytes.decode('utf-8')
    
    soup = BeautifulSoup(webpage_text, 'lxml')
    
    houselist = soup.find_all("li",class_="zl-card")
    
    for item in houselist:
        for link in item.find_all("a", class_="zl-card-inner" , href=True):
            houselinks.append(base_url + link["href"])

print(len(houselinks))
```

```python
datalist1 = []

for link in houselinks:
    req = Request(url=link, headers={'User-Agent': 'Mozilla/5.0'})
        
    # Open the URL and read the response
    response1 = urlopen(req)
    webpage_bytes1 = response1.read()
    webpage_text1 = webpage_bytes1.decode('utf-8')
    
    
    soup1 = BeautifulSoup(webpage_text1, 'lxml')

    try:
        name = soup1.find('h1').text.strip()
    except:
        name = 0
    try:
        price = soup1.find('div', class_="col-md-3 price-info-text").text.strip().replace(".","").replace(" TL", "")
    except:
        price = 0
    try:
        isim = soup1.find("div", {"class":"col-xs-12"}).h1.text
    except:
        isim = 0
    try:
        net_m2 = soup1.find("ul", {"class":"row attribute-detail-list"}).findAll("li",{"class":"col-md-6"})[2].find("span").text.replace("m²","").strip()
    except:
        net_m2 = 0
    try:
        brut_m2 = soup1.find("ul", {"class":"row attribute-detail-list"}).findAll("li",{"class":"col-md-6"})[3].find("span").text.replace("m²","").strip()
    except:
        brut_m2 = 0
    try:
        oda_salon = soup1.find("ul", {"class":"row attribute-detail-list"}).findAll("li",{"class":"col-md-6"})[4].find("span").text.strip()
    except:
        oda_salon = 0
    try:
        banyo_sayisi = soup1.find("ul", {"class":"row attribute-detail-list"}).findAll("li",{"class":"col-md-6"})[5].find("span").text.strip()
    except:
        banyo_sayisi = 0
    try:
        bina_yasi = soup1.find("ul", {"class":"row attribute-detail-list"}).findAll("li",{"class":"col-md-6"})[9].find("span").text.strip()
    except:
        bina_yasi = 0
    try:
        fotograf_sayisi = soup1.find("div", {"class":"detail-images-slide-container hidden-xs hidden-sm"}).a.span.text.replace("+","")
    except:
        fotograf_sayisi = 0
    
    data = {
        'name': name,
        'price': price,
        'isim': isim,
        'netmetrekare': net_m2,
        'brutmetrekare': brut_m2,
        'odasalon': oda_salon,
        'banyosayisi': banyo_sayisi,
        'binayasi': bina_yasi,
        'fotografsayisi': fotograf_sayisi
        }
    datalist1.append(data)
    print("Saving", data["name"])

df = pd.DataFrame(datalist1)
print(df.head())

```
