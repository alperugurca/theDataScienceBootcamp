```python
footers = soup2.find_all("footer")
```

```python
footers

prices = []

for footer in footers:
    text = footer.get_text(separator=' ').strip()  # Get text and clean it up
    lines = text.split('Son Güncelleme Tarihi:')
    
    if len(lines) == 2:  # Ensure there are both price and date
        price = lines[0].strip()  # Extract the price
        prices.append(price)

# Create a DataFrame from the extracted data
df = pd.DataFrame({
    'Price': prices,
})

print(df)
```

```python
df['Price'] = df['Price'].astype(str)
```

```python
df["Price"] = df["Price"].str.lower()
```

```python
df["Price"] = df["Price"].str.replace(".","")
```

```python
df["Price"] = df["Price"].str.replace(" ","")
df["price"] = df["Price"].str.replace("tl","")
```

```python
listings = soup.find_all('div', class_='item-content')
```

```python
listings
```

```python
data = []

for listing in listings:
    property_title = listing.find('a', class_='classified-title').text.strip()
    location = listing.find('small', class_='property-categories').find('span').text.strip()
    property_type = listing.find('div', class_='item-properties').find_all('span')[0].text.strip()
    apartment_details = listing.find('div', class_='item-properties').find_all('span')[1].text.strip()
    price = listing.find('footer').find('strong').text.strip()
    update_date = listing.find('footer').find('span').text.strip()

    
    data.append({
        'Property Title': property_title,
        'Location': location,
        'Property Type': property_type,
        'Apartment Details': apartment_details,
        'Price': price,
        'Update Date': update_date
    })


df = pd.DataFrame(data)

print(df)
```
