# Regression

```python
import pandas as pd
from pycaret.regression import *

df=pd.read_pickle('kc_house.pkl')

setup(data=df, target='price')
```

```python
best_model=compare_models()
```

# Classification

```python
df = pd.read_csv('pima-indians-diabetes.csv')

from pycaret.classification import *

setup(data=df, target='Outcome')
```

```python
compare_models()
```

```python
best_model=create_model('lr')
```

```python
plot_model(best_model, plot='feature') # hangi sütun daha önemli
```

# Clustering

```python
df=pd.read_excel('cars.xls')
from pycaret.clustering import *
setup(data=df)
```

```python
kmeans=create_model('kmeans', 4)
```

```python
plot_model(kmeans, plot='tsne')
```

```python
plot_model(kmeans)
```

```python
plot_model(kmeans, plot='elbow') # kaça ayıracağımıza karar verme
```
