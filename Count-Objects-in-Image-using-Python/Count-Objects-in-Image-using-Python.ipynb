# Count Objects in an Image using Python

Let's explore how to use the cvlib library to count the number of objects in an image using Python.

Example Solution: https://thecleverprogrammer.com/2021/05/11/count-objects-in-image-using-python/

Hugging Face: https://huggingface.co/spaces/alperugurcan/object-detection-yolov5



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox

# Read the image file
image = cv2.imread('1.jpg')

# Use cvlib to detect objects in the image
bbox, label, conf = cv.detect_common_objects(image)
```

```python
# Draw bounding boxes on the image
output_image = draw_bbox(image, bbox, label, conf)

# Convert BGR to RGB for correct display with matplotlib
output_image = cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)

# Display the output image
plt.figure(figsize=(12, 8))
plt.imshow(output_image)
plt.axis('off')
plt.title("Object Detection Result")
plt.show()

# Count and display all unique labels
label_counts = {}
for item in label:
    if item in label_counts:
        label_counts[item] += 1
    else:
        label_counts[item] = 1

print("Detected objects and their counts:")
for item, count in label_counts.items():
    print(f"{item}: {count}")

print("\nTotal number of objects detected:", len(label))
print("Unique labels detected:", list(label_counts.keys()))
```
