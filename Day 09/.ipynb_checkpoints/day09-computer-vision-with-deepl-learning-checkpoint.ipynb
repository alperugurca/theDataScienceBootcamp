# Computer Vision with Deep Learning

```python
import tensorflow as tf
```

```python
mnist=tf.keras.datasets.mnist # El yazısı rakamları tanıma
```

```python
(train_images, train_labels), (test_images,test_labels) = mnist.load_data()
```

### Normalize Data, 255 e bölerek 0 ve 1 arasına normalize etmek

```python
train_images=train_images/255.0
test_images=test_images/255.0
```

```python
0/255.0
```

```python
plt.imshow(train_images[54], cmap="gray")
```

### Convolutional Neural Network - CNNs Image Classification

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Conv2D, InputLayer, Reshape, MaxPooling2D, Flatten
```

```python
model=Sequential()
model.add(InputLayer(input_shape=(28,28)))
model.add(Reshape(target_shape=(28,28,1))) # her pikselde 1 numara var
# filters her bir büyük kare
model.add(Conv2D(filters=12, kernel_size = (3,3), activation="relu")) # Conv2D, LSTM, GAN(Yeni resim), 3x3 alır
model.add(MaxPooling2D(pool_size=(2,2))) # 2x2 al max al
model.add(Conv2D(64, (3,3), activation="relu"))
model.add(Flatten()) # 1 boyutlu sütun haline getirir
model.add(Dense(10)) # 10 farklı cevap var
model.compile(optimizer="adam", loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=["accuracy"])
```

```python
model.fit(train_images, train_labels, validation_split=0.1, epochs=10)
```

```python
loss, accuracy = model.evaluate(test_images, test_labels)
```

```python
accuracy 
```

```python
model.save("ElYazisiRakamlariTahmin.h5")
```

### Deep Learningde Başarı oranını artırma teknikleri
1. Layers sayısını artır. model.add(Dense(60))

2. Neuron sayısını artır. model.add(Dense(9999999999))

3. Dropout ile Overfitting Engelleme, daha iyi öğrenme sağlama. 
* from tensorflow.keras.layers import Dropout
* model.add(Dropout(0.25)) , gelen verinin %25ini at

4. Batch Normalization ile verileri normalleme
* from tensorflow.keras.layers import BatchNormalization
* model.add(BatchNormalization())

```python
model=Sequential()

model.add(InputLayer(input_shape=(28,28)))
model.add(Reshape(target_shape=(28,28,1))) # her pikselde 1 numara var

model.add(Conv2D(filters=12, kernel_size = (3,3), activation="relu")) # Conv2D, LSTM, GAN(Yeni resim), 3x3 alır, # filters her bir büyük kare
model.add(MaxPooling2D(pool_size=(2,2))) # 2x2 al max al
model.add(Conv2D(64, (3,3), activation="relu"))
model.add(Flatten()) # 1 boyutlu sütun haline getirir
model.add(BatchNormalization()) ### Kullanılmayan verileri temizle tarzı bir şey
model.add(Dense(60)) ### layer +1 ve neuron +60 ekledik Dense(Neuron_number)
model.add(Dropout(0.25)) ### Dropout
model.add(Dense(10)) # 10 farklı cevap var
model.compile(optimizer="adam", loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=["accuracy"])
```
