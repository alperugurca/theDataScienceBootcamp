# Imputation Techniques

```python
import pandas as pd
```

```python
df = pd.read_csv('ExampleData.csv')
```

```python
df
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
# bo≈ü verileri linear reggession ile doldurma(Rakam)
lr=LinearRegression()
imp=IterativeImputer(estimator=lr) #
imputed_data=imp.fit_transform(df) #
```

```python
pd.DataFrame(imputed_data) # dataframe ≈üeklinde g√∂ster
```

```python
# KNN Imputer = En yakƒ±n kom≈üu deƒüer ile doldur
```

```python
from sklearn.impute import KNNImputer
```

```python
knn = KNNImputer()
```

```python
imputed_data2 = knn.fit_transform(df)
```

```python
imputed_data2
```

```python
pd.DataFrame(imputed_data2, columns=[df.columns]) # columns geri getirme
```

```python
# Ortalama veya orta deƒüer ile doldurma
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(strategy='mean') # mean ile doldur ortalama - median,mode
```

```python
imputed_data3 = imp.fit_transform(df)
```

```python
type(imputed_data3)
```

```python
pd.DataFrame(imputed_data3,columns=[df.columns]) # columns geri getirme
```
