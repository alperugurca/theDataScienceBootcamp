# Linear Regression ile veri doldur (numeric)

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression() # lryi ata
```

```python
imp = IterativeImputer(estimator = lr)
```

```python
imputed_data = imp.fit_transform(df) # df olmaktan çıktı
```

```python
pd.DataFrame(imputed_data, columns=[df.columns]) # veriler doldu
```

# KNN ile doldur

```python
from sklearn.impute import KNNImputer # KNN ile doldur en yakın komşu
```

```python
knn = KNNImputer()
```

```python
imputed_data2 = knn.fit_transform(df)
```

```python
imputed_data2 = pd.DataFrame(imputed_data2, columns=[df.columns]) # sütun isimlerini geri getir
```
