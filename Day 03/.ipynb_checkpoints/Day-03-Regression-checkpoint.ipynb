## Feature Engineering

```python
df.corr(numeric_only=True)
```

```python
abs(df.corr(numeric_only=True)["Price"].sort_values(ascending=False)) # korelasyon sıralaması
```

```python
x=df[["Mileage", "Cylinder", "Liter", "Cruise"]]
y=df[["Price"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
X_train.shape, X_test.shape
```

## Modelling

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression() # kolay erişim
```

```python
model=lr.fit(X_train, y_train) # makine öğrenmesi fitting
```

```python
tahmin=model.predict(X_test) # X_test dahasındaki verileri tahmin et
```

```python
from sklearn.metrics import r2_score, mean_squared_error # analiz araçları yükledik
```

```python
mean_squared_error(y_test, tahmin)
```

```python
r2_score(y_test, tahmin)
```

```python
# Label encoding 1==ferrari 2==tesla birinin diğerine üstünlüğü olmadığından hatalı olur. # sparse matrix == yazıları numeric değerleri çevirme tekniği
```

```python
x=df[["Mileage", "Cylinder", "Liter", "Cruise", "Make", "Trim"]]
y=df[["Price"]]
```

```python
x.head()
```

```python
x=pd.get_dummies(x,drop_first=True) # bir değişkeni ayırarak diğerlerine bakarak bir değişkeni otomatik tahmin ediyor. 
```

```python
x.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model=lr.fit(X_train, y_train) # makine öğrenmesi fitting
```

```python
tahmin=model.predict(X_test) # X_test dahasındaki verileri tahmin et
```

```python
mean_squared_error(y_test, tahmin)**0.5 # karekökünü aldık # root mean square error == kısaca hata # 1523 dolar sapma # -lerden kurtul veya mutlak değer
```

```python
r2_score(y_test, tahmin) # %97 doğruluk
```

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet # Regularization Ridge= hataya focus, Lasso= Katsayılara focus, ElasticNet= ikisini de kullan hangisi iyiyse onu seçmek için
```

```python
R=Ridge()
```

```python
L=Lasso()
```

```python
E=ElasticNet()
```

```python
Rmodel=R.fit(X_train,y_train) # eğit
```

```python
tahmin2=Rmodel.predict(X_test)
```

```python
mean_squared_error(y_test, tahmin)**0.5
```

```python
r2_score(y_test, tahmin2)
```

```python
Lmodel=L.fit(X_train,y_train) # eğit
```

```python
tahmin3=Lmodel.predict(X_test)
```

```python
mean_squared_error(y_test, tahmin3)**0.5
```

```python
r2_score(y_test, tahmin3)
```

```python
Emodel=E.fit(X_train,y_train)
```

```python
tahmin4=E.predict(X_test)
```

```python
mean_squared_error(y_test, tahmin4)**0.5
```

```python
r2_score(y_test, tahmin4)
```

```python
# Coefficiants - Katsayılar
```

```python
lr.coef_
```

```python
feature_importance=pd.DataFrame({"Feature":X_train.columns, "Coefs":lr.coef_[0]})
```

```python
feature_importance=feature_importance.sort_values(by="Coefs", ascending=False)
```

```python
feature_importance # katsayı alarak önemleri görme 0'a yaklaştıkça önemsiz
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,15))
plt.barh(feature_importance["Feature"],feature_importance["Coefs"]) # feature önemlerini gör
plt.gca().invert_yaxis()
```

```python
# Regression başarı oranı için
# 1. R2_score > 0.85
# 2. Dusuk RME istiyoruz
# 3. Residuals should follow normal distiribution ## hata ile gerçek veri arası fark
# 3.1 tahmin-y_test
```
