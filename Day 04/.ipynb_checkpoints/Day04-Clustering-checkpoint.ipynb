# Clustering: Veriyi Benzerliklerine Göre Kümelendirme

```python
import pandas as pd
```

```python
df=pd.read_excel("cars.xls")
```

```python
df.head()
```

Hatırlatma
Regression:: r2, RMSE
Classification Accuracy:: F1 Score, Classification Error and Confusion Matrix
Clustering Silhouette Score:: Yüksek olması gerekmez

```python
x=df[["Price","Mileage","Cylinder"]]
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
x.head()
```

```python
model = KMeans(7)
```

```python
model = model.fit(x)
```

```python
tahmin = model.predict(x)
```

```python
tahmin
```

```python
x["cluster"]=tahmin
```

```python
x.head()
```

```python
silhouette_score(x,tahmin)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x.Mileage,x.Price, c=tahmin)
plt.colorbar()
plt.xlabel("km")
plt.ylabel("fiyat")
```

```python
#wcss= within cluster sum of squares
wcss=[]
ss=[]
for i in range(2,10):
    model = KMeans(i)
    model = model.fit(x)
    tahmin=model.predict(x)
    ss1=silhouette_score(x,tahmin)
    ss.append(ss1)
    print(ss1)
    wcss.append(model.inertia_)
```

```python
plt.plot(range(2,10),wcss)
```

```python
import seaborn as sns
```

```python
sns.lineplot(x=range(2,10),y=wcss, marker="*")
```

```python
data = linkage(x, method="ward", metric="euclidean")
```
