```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
df1=pd.read_csv("iris.csv")
```

```python
print(df1.head())

X = df1[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df1['species']
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y) # transform ile türleri 0 1 2 olarak ayır
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train) # modeli eğit
```

```python
def predict_species(row):

    row_data = row.values.reshape(1, -1)
    prediction = model.predict(row_data)[0]
    species_name = label_encoder.inverse_transform([prediction])[0]
    return species_name
```

```python
def get_user_input_and_predict():
    print("Please enter the following measurements:")
    
    sepal_length = float(input("Sepal length: "))
    sepal_width = float(input("Sepal width: "))
    petal_length = float(input("Petal length: "))
    petal_width = float(input("Petal width: "))
    
    # Create a pandas Series with the input data
    new_data = pd.Series({
        'sepal_length': sepal_length,
        'sepal_width': sepal_width,
        'petal_length': petal_length,
        'petal_width': petal_width
    })
    
    # Predict species
    predicted_species = predict_species(new_data)
    print(f"The predicted species is: {predicted_species}")

# Run the function to get user input and predict species
get_user_input_and_predict()
```
