# Popularity-based

Dataset: https://grouplens.org/datasets/movielens/100k/

```python
import pandas as pd
import numpy as np
```

```python
rcolumns = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', header=None, usecols=range(3), names=rcolumns)
ratings.head()
```

```python
movies=pd.read_csv('ml-100k/u.item', encoding='latin-1', sep='|', header=None, names=['movie_id', 'title'], usecols=[0,1], engine='python')
movies.head()
```

```python
user=pd.read_csv('ml-100k/u.user', sep='|', header=None, names=['user_id', 'age', 'gender', 'occupation', 'zip_code'])
user
```

```python
movie_grouped=ratings.groupby('title').agg({'rating': [np.size, np.sum,np.mean]})
```

```python
movie_grouped
```

```python
popular_movies=movie_grouped.sort_values([('rating', 'mean')], ascending=False)
```

```python
popular_movies.head(10)
```

```python
grouped_sum=movie_grouped['rating']['sum'].sum()
```

```python
popular_movies['percentage']=movie_grouped['rating']['sum']/grouped_sum*100

popular_movies.head(10)
```

```python
popular_movies.sort_values('percentage', ascending=False)
```

```python
popular_movies['percentage'].rank(ascending=False)
```
