# Emotion Detection from Photos: A Simple Application Using FER

Example Solution: https://thecleverprogrammer.com/2020/08/16/emotion-detection-model/

Hugging Face: https://huggingface.co/spaces/alperugurcan/Emotion-Detector

```python
# Emotion_Detection.ipynb

import cv2
import numpy as np
from fer import FER
from PIL import Image
import matplotlib.pyplot as plt

# Initialize the FER model
detector = FER(mtcnn=True)

def detect_emotion(image_path):
    # Read the image
    img = cv2.imread(image_path)

    # Detect emotions
    emotions = detector.detect_emotions(img)[0]

    # Get the dominant emotion
    dominant_emotion = max(emotions['emotions'], key=emotions['emotions'].get)

    # Display the image with emotion
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(f"Detected Emotion: {dominant_emotion}")
    plt.axis('off')
    plt.show()

    return dominant_emotion

# Test the function
test_image = "/content/man.jpg"
result = detect_emotion(test_image)
print(f"Detected emotion: {result}")
```
