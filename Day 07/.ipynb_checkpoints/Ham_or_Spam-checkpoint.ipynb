```python
#from imblearn.over_sampling import SMOTE # imbalanced verileri dengeleme
#x,y = SMOTE().fit_resample(x,y)
```

```python
# temizleme işlemi
df['text'] = df['text'].str.lower()
df['text'] = df['text'].str.replace('[^\w\s]', '', regex=True)
df['text'] = df['text'].str.replace('\d+', '', regex=True)
df['text'] = df['text'].str.replace('\n', '', regex=True)
df['text'] = df['text'].str.replace('\r', '', regex=True)
```

```python
x=df['text']
y=df['class']
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# pipeline == Yapılacak işlemleri sıraya koyma işlemi
from sklearn.pipeline import Pipeline
```

```python
# Cevabı evet hayır olduğu için classifier
clf=Pipeline([('tfidf',TfidfVectorizer()),('classifier', RandomForestClassifier())])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.20,random_state=42)
```

```python
clf.fit(x_train,y_train) # Pipeline çalıştı önce tfid sonra rfc
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred) # kullan
```

```python
from sklearn.svm import SVC # Support Vector Machine, ekstra boyut ekleme
```

```python
svm=Pipeline([('tfidf',TfidfVectorizer()),('classifier', SVC())])
```

```python
svm.fit(x_train,y_train)
```

```python
y_pred=svm.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred) # kullan
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

```python
ham=df[df['class'] =='ham']
spam=df[df['class'] == 'spam']
```

```python
# Data augmentation - Varolan datadan yeni datalar oluşturma
# Synthetic data - Sentetik data
```

```python
#!pip install wordcloud
from wordcloud import WordCloud
from wordcloud import STOPWORDS
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
' '.join(ham['text'])
```

```python
def wc(data,color): # wordcloudddddd
    plt.figure(figsize=(10,10))
    mask=np.array(Image.open('cloud.png')) # çerçeve ekleme
    wc=WordCloud(background_color=color,stopwords=STOPWORDS,mask=mask)
    wc.generate(' '.join(data))
    plt.imshow(wc)
    plt.axis('off')
```

```python
wc(ham['text'], 'white')
```

```python
wc(spam['text'],'red')
```

```python
spam = spam.sample(ham.shape[0], replace=True) # under
```

```python
spam.shape, ham.shape
```

```python
data=pd.concat([ham,spam],ignore_index=True)
```

```python
data.shape
```

```python
data.head()
```

```python
sns.countplot(x=data['class'],hue=data['class'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(data['text'],data['class'], test_size=0.20,random_state=42)
```

```python
clf=Pipeline([('tfidf',TfidfVectorizer()),('classifier', RandomForestClassifier())])
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
nlp = spacy.load(name='en_core_web_md')
```

```python
s1 = '''As if nothing can warn
The silence within us
No words, no words, nothing
Bring the eyes, bring the eyes.
Nothing else, so we agree
When a leaf touches another leaf
So close, so docile
Bring the hands, bring the hands
I'm saying, love today is being against something
Let us unite and release two single shadows. '''
```

```python
print(s1.split()) # Tokenize ettik
```

```python
doc1 = nlp(s1)
```

```python
#for token in doc1:
    #print(token)
s1 = [x for x in doc1]
print(s1)
```

```python
[x.lemma_ for x in s1] # Kök bölma lemma
```

```python
doc = nlp('Apple is looking at buying U.K. startup for $1 billion')
```

```python
[x.lemma_ for x in doc]
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_,token.tag_)
```

```python
from spacy import displacy
```

```python
displacy.render(doc,style='dep')
```
